```python
import numpy as np
import cv2
from skimage import measure
```

```python
import matplotlib.pyplot as plt
```

```python
source = cv2.imread("C:\\Users\\a\\Desktop\\assign\\3.png")
target = cv2.imread("C:\\Users\\a\\Desktop\\assign\\4.png")
```

```python
print(source.shape)
print(target.shape)
```

```python
dim = (500,500) 
sour = cv2.resize(source,dim)
tar = cv2.resize(target,dim)
```

```python
def compare_images(imageA, imageB, title):
     
    s = measure.compare_ssim(imageA, imageB)
 
    fig = plt.figure(title)
    plt.suptitle("SSIM: %.2f" % (s))
 
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")
 
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis("off")
 
    plt.show()
```

```python
original = cv2.cvtColor(sour, cv2.COLOR_BGR2GRAY)
contrast = cv2.cvtColor(tar, cv2.COLOR_BGR2GRAY)
```

```python
compare_images(original, original, "Original vs. Original")
compare_images(original, contrast, "Original vs. Contrast")
```
