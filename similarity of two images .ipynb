```python
import numpy as np
import cv2
from skimage import measure
import matplotlib.pyplot as plt
```

```python
source = cv2.imread('C:\\Users\\hem\\Desktop\\similarity\\sim1.jpg')
target = cv2.imread('C:\\Users\\hem\\Desktop\\similarity\\sim2.jpg')
print(source.shape, target.shape)
dim = (500,500) 
sour = cv2.resize(source,dim)
tar = cv2.resize(target,dim)
print("resized to Dimension")
print(sour.shape,tar.shape)
```

```python
def compare_images(imageA, imageB, title):
     
    s = measure.compare_ssim(imageA, imageB)
 
    fig = plt.figure(title)
    plt.suptitle("SSIM: %.2f " % (s*100))
 
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")
 
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis("off")
 
    plt.show()
    return s
```

```python
original = cv2.cvtColor(sour, cv2.COLOR_BGR2GRAY)
contrast = cv2.cvtColor(tar, cv2.COLOR_BGR2GRAY)
compare_images(original, contrast, "Original vs. Contrast")
print("we concluded 84% of similarity between this two images by using ssim method")
```

# finding similarity between two images  using brute force matching technique using that difference

```python
import cv2
import matplotlib.pyplot as plt
orb = cv2.ORB_create()
reshape = (800,800)
```

# cropped with this two graph 

```python
img1 = cv2.imread('C:\\Users\\hem\\Desktop\\similarity\\sim1.jpg')          
img2 = cv2.imread('C:\\Users\\hem\\Desktop\\similarity\\sim2.jpg')
resha = (600,600)
img1 = cv2.resize(img1,resha)
img2 = cv2.resize(img2,resha)
```

```python
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
print("Number of matches or similarity identified b/w 2 images ",len(matches))
```

```python
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:50],None, flags=2)
fig=plt.figure(figsize=(15, 15), dpi= 80, facecolor='w', edgecolor='k')
fig.suptitle('shows what are all the  similarity between two images ', fontsize=20)
plt.imshow(img3)
plt.show()
```
